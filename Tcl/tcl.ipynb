### TCL语言
Tcl是Tool Command Language缩写形式。加州大学伯克利分校的John Ousterhout设计了它。它是脚本语言和自己的解释器的组合，嵌入到应用程序中，我们用它开发。

Tcl最初是为Unix开发的。然后将其移植到Windows，DOS，OS / 2和Mac OSX。Tcl与Bourne Shell（Sh），C Shell（csh），Korn Shell（sh）和Perl等其他unix shell语言非常相似。
它旨在提供程序与其他程序进行交互的能力，还可以作为嵌入式解释器。尽管如此，最初的目标是让程序进行交互。

Tcl是一种工具命令语言，命令是语言中最重要的部分。Tcl命令内置在语言中，每个都有自己的预定义功能。这些命令形成语言的保留字，不能用于其他变量命名。这些Tcl命令的优点是，您可以为任何这些命令定义自己的实现来替换原始的内置功能。\
每个Tcl命令验证输入，并减少了解释器的工作。\
Tcl命令实际上是单词列表，第一个单词表示要执行的命令。下一个字代表论点。为了将单词分组成单个参数，我们用“”或{}括起多个单词。\
Tcl命令的语法如下：
```tcl
commandName argument1 argument2 ... argumentN
```
我们来看一个简单的Tcl命令示例
```tcl
#!/usr/bin/tclsh
puts "Hello, world!"
```
当执行上述代码时，输出结果如下
```shell
Hello, world!
```
在上面的代码中，“puts”是Tcl命令，“Hello World”是参数1。如前所述，我们用“”来组合两个字。
我们来看看另外一个Tcl命令的例子，有两个参数
```tcl
#!/usr/bin/tclsh
puts stdout "Hello, world!"
```
当执行上述代码时，输出结果如下
```shell
Hello, world!
```
在上面的代码中，"puts"是Tcl命令，"stdout"是argument1，“Hello World”是argument2。这里，stdout使程序在标准输出设备中打印。

### 注释

注释就像在Tcl程序中帮助文本，而解释器忽略它们。注释可以使用开头的hash _（＃）标记来写。
```tcl
#!/usr/bin/tclsh
# my first program in Tcl
puts "Hello World!" 
```

当执行上述代码时，输出结果如下
```shell
Hello World!
```
内联注释使用;＃。下面给出了一个例子。
```tcl
#!/usr/bin/tclsh
puts "Hello World!" ;# my first print in Tcl program
```
当执行上述代码时，输出结果如下
```shell
Hello World!
```

### 变量替代
在变量替换中，$在变量名称之前使用，并返回变量的内容。将值设置为变量并打印的简单示例如下所示

```tcl
#!/usr/bin/tclsh
set a 3
puts $a
```

### Tcl 数据类型
Tcl的原始数据类型是字符string，通常我们可以在Tcl上找到引用字符string的语言。\
这些原始数据类型又为列表和关联数组创建复合数据类型。
```tcl
    set myVariable 18
    puts $myVariable
```
上述语句将创建一个变量名myVariable并将其存储为字符string，即使我们还没有使用双引号。现在，如果我们尝试对变量进行算术，它将自动转换为一个整数。一个简单的例子如下所示：
```tcl
    set myVariable 18
    puts [expr $myVariable + 6 + 9]
```
当执行上述代码时，输出结果如下
```shell
    33
```
需要注意的一点是，这些变量没有任何默认值，在使用之前必须赋值。
如果我们尝试使用put打印，则该数字将转换为正确的字符string。在内部和外部有两个表示，帮助Tcl轻松地与其他语言相比创建复杂的数据结构。此外，由于其动态对象性质，Tcl更加高效。

#### 字符string表示

与其他语言不同，在Tcl中，当单个单词不需要双引号。一个例子可以是
```tcl
    set myVariable hello
    puts $myVariable
```
当执行上述代码时，输出结果如下
```shell
    hello
```
当我们要表示多个字符string时，我们可以使用双引号或大括号。如下所示
```tcl
    set myVariable "hello world"
    puts $myVariable
    set myVariable {hello world}
    puts $myVariable
```
当执行上述代码时，输出结果如下
```shell
    hello world
    hello world
```
#### 列表
列表只不过是一组元素。可以使用一组使用双引号或大括号的单词来表示一个简单的列表。一个简单的列表如下所示：
```tcl
    #!/usr/bin/tclsh
    set myVariable {red green blue}
    puts [lindex $myVariable 2]
    set myVariable "red green blue"
    puts [lindex $myVariable 1]
```
当执行上述代码时，输出结果如下
```shell
    blue
    green
```
#### 关联数组
关联数组具有不一定是整数的索引（键）。它通常是一个像键值对一样的字符string。一个简单的例子如下所示：
```tcl
    #!/usr/bin/tclsh
    set  marks(english) 80
    puts $marks(english)
    set  marks(mathematics) 90
    puts $marks(mathematics)
```
当执行上述代码时，输出结果如下
```shell
    80
    90
```

Tcl手柄通常用于表示文件和图形对象。这些可以包括网络请求的处理以及string行端口通信，套接字或I / O设备等其他通道。以下是创建文件句柄的示例。
```tcl
    set myfile [open "filename" r]
```
